```python
#dividing features into dependent and independent
X=df1.drop('FWI',axis=1)
y=df['FWI']
```

```python
X
```

```python
y
```

```python
##train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
##feature selection based on correleation
##check for multicolinearity
plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
```

```python
def correlation(dataset,threshold):
  col_corr=set()
  corr_matrix=dataset.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if abs(corr_matrix.iloc[i,j]>threshold):
        colname=corr_matrix.columns[i]
        col_corr.add(colname)
  return col_corr
```

```python
##setting threshold--in company domain expert set it
corr_features=correlation(X_train,0.85)
```

```python
##drop features when correlation is morethan .85
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

```python
#feature scaling --standardization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

BOX plot to understand effect of Standard Scaler

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title("X_train after scaling")
```

```python
#Model Training
#Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
linreg.fit(X_train_scaled,y_train)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
y_pred=linreg.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("mae:",mae)
print("r2_score:",score)
```

```python
plt.scatter(y_test,y_pred)
```

#Lasso Regression

```python
##lasso Regression
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso=Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred=lasso.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("mae:",mae)
print("r2_score:",score)
```

```python
plt.scatter(y_test,y_pred)
```

Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
ridge=Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred=ridge.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("mae:",mae)
print("r2_score:",score)
```
