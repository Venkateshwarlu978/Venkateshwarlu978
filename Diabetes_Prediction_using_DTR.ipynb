```python
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
```

```python
print(dataset['DESCR'])
```

```python
dataset
```

```python
import pandas as pd
df_diabetes=pd.DataFrame(dataset.data,columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
```

```python
df_diabetes.head()
```

```python
X=df_diabetes
y=dataset['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(X_train,y_train)
```

hyperparamter Tunning


```python
param={'criterion':['squared_error','friedman_mse','absolute_error'],
       'splitter':['best','random'],
       'max_depth':[1,2,3,4,5,10,15,20,25],
       'max_features':['auto','sqrt','log2']}
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=DecisionTreeRegressor(),param_grid=param,cv=5,scoring='neg_mean_squared_error')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
rs=r2_score(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
print(rs)
print(mse)
print(mae)
```

```python
selectmodel=DecisionTreeRegressor(criterion='squared_error',
 max_depth=2,
 max_features='log2',
 splitter='best')
```

```python
selectmodel.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(selectmodel,filled=True)
```
